# Máquinas y Mecanismos

[Tipos de mecanismos](https://adeanstudio.wordpress.com/2018/11/16/tipos-de-mecanismos/)

Las **máquinas** se diseñan y construyen para facilitar las tareas a la hora de realizar un trabajo o aplicar una fuerza. Sus partes son: la **estructura** y los **mecanismos**.

Los **mecanismos**   transmitenir y transforman energía y movimiento desde un elemento motriz(motor) a un elemento receptor modificando sus características.

## Clasificación de los mecanismos.
Según la función que realizan se clasifican en:

* **Mecanismos de transmisión del movimimiento**. Transmiten el mismo tipo de movimiento desde el elemento motor al receptor. Este movimiento puede ser lineal o circular.

* **Mecanismos de transformación del movimeinto**. Transformana el movimiento del elemento motor en otro tipo de movimiento en el elemento receptor. Movimiento circular en lineal, movimiento circular en movimiento alternativo.


## Mecanismos de transmisión lineal.

### La palanca.
Está formada por una barra rígida y un punto de apoyo.



**Ley de la palanca:** Cuando la palanca se encuentra en equilibrio se cumple que el producto de la fuerza por su brazo es igual a producto de la resistencia por su brazo.




* **Brazo**. b es la distancia entre el punto de aplicación de las fuerzas y el punto de apoyo. $b_R$ es el brazo de resistencia y $b_F$ el brazo de potencia.
* **Resistencia o carga.** R es la fuerza que se vence o se obtiene como resultado de utilizar la palanca.
* **Punto de apoyo.** PA es el punto sobre el que gira la palanca o sobre el que se apoya.
* **Fuerza.** F es la fuerza que se aplica en un extremo de la palanca.

**Palanca de primer género:** El punto de apoyo está entre la fuerza y la resistencia.

 

**Palanca de segundo grado:** Se obtiene cuando colocamos la resistencia entre la potencia y el punto de apoyo

 

**Palanca de tercer grado:** Se obtiene cuando ejercemos la potencia entre el punto de apoyo y la resistencia.

 
 
 
 
 

### La polea.
La polea es una rueda ranurada que gira alrededor de un eje que se halla sujeto a una superficie fija. Por la ranura de la polea pasa la cuerda.

Si queremos levantar una carga, podemos tirar de ella hacia arriba; pero esto puede resultar muy incómodo. Para evitarlo, se utiliza la polea fija, la polea móvil o el polipasto.

**Polea Fija**



Altirar de la cuerda, la polea gira sobre su eje y se levanta la carga.  la fuerza que tenemos que aplicar para elevar la carga es igual al peso que queremos levantar.

**Polea móvil**



la fuerza que tenemos que aplicar para elevar una carga con una polea móvil es la mitad de la fuerza resistente de esa carga. $$F = \frac {R}{2}$$

**Polipasto**



Combinando varias poleas fijas y móviles, ejerciendo una fuerza pequeña podremos levantar un gran peso.  

$$F = \frac {R}{2^n}$$

Donde **n** es el número de poleas móviles


## Mecanismos de transmisión circular.

### Rueda  
Objeto circular, que puede girar sobre un eje y tiene multiples aplicaciones. Tambien se conoce como maquina simple.



### Sistemas de polea y correa.
 Este sistema está formado por dos poleas unidas mediante una correa, de tal forma que el movimiento circular se transmite de una polea a la otra.
 
 

La ecuación que relaciona las velocidades de giro de las poleas y sus diámetros es:

$$d_A · \omega_A = d_B · \omega_B$$

La velocidad de giro se puede modificar cambiando la relación de tamaño de las poleas.




**La velocidad se multiplica**  cuando la polea motriz es más grande que la polea conducida. 

**La velocidad se mantiene** cuando las dos poleas, motriz y conducida, tienen el mismo tamaño. 

**La velocidad se reduce**  cuando la polea motriz es más pequeña que la polea conducida. 


### Transmisión por cadenas
Es uno de los metodos mas comunes de transmision. Esta formado por un piñon que atrae y otro que arrastra, mediante un sistema dentado.



## Mecanismos de transformación del movimeinto.


### Sistema biela-manivela
El sistema biela-manivela transforma el movimiento circular de una manivela en un movimiento lineal o alternativo, y viceversa, al estar conectada a una barra rígida o biela.




### Sistema de piñón y cremallera.
Piñon cremallera:es un mecanimso compuesto por un piñon  o rueda dentada de dientes rectos, que engrana con una barra dentada denominada cremallera  de forma que, cuando el piñón gira, la barra dentada se desplaza longitudnalmente.



### Tornillo y tuerca:
El tornillo es una pieza mecanica cilindrica o conica, con un resalte helecoidal que la recorre o parcialmente una cabeza con una ranura para alujarla pala del destornillador. tuerca: pieza de metal de cuatro o seis ladoscon un agujero en el centro, se ajusta a la rosca de un tornillo para fijarlo.



### Engranajes
los engranagues son juegos de ruedas que disponen de unos elementos salientes denominados dientes que encajan entre si.



### Cigüeñal 
Parte del motor del automóvil y otras máquinas que consiste en un eje con varios codos, en cada uno de los cuales se ajusta una biela, y está destinada a transformar el movimiento rectilíneo de los pistones en rotativo, o viceversa.



### Leva  
Pieza que gira alrededor de un punto que no es su centro, transformando el movimiento circular continuo en otro rectilíneo alternativo.



### Excéntrica  
Pieza circular de una máquina, cuyo eje de rotación no ocupa el centro geométrico y que está destinada a transformar un movimiento de rotación en uno de otra clase, especialmente rectilíneo.
