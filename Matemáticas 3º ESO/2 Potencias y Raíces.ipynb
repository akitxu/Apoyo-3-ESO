## Potencias con exponente entero positivo

<font color = "blue">
* $a^b = \underbrace{a\cdots a}_{b \; \text{veces}}$

* $-a^b = -(\underbrace{a\cdots a}_{b \; \text{veces}})$

* $(-a)^b = \underbrace{(-a)\cdots (-a)}_{b \; \text{veces}}$

 </font>

<font color = "blue">
**Propiedades** Las potencias con exponente entero positivo cumplen las siguientes propiedades.

* **Potencias con la misma base**

 $$ a^m \cdot a^n = a^{m+n}$$
 
 $$ \frac {a^m} {a^n} = a^{m-n} \quad si \quad a \neq 0 \quad y \quad \; si \; m > n$$
    </font>

<font color = "blue">
**Potencias con el mismo exponente**

$$a^n \cdot b^n =(a \cdot b)^n$$

$$ \frac {a^n}{b^n} = (\frac {a}{b})^n \quad si \quad b \neq  0$$
</font>

<font color = "blue">
**Potencia de potencia**

$$ (a^m)^n = a^{m \cdot n}$$
    </font>

<font color = "blue">
## Potencias con exponente 0

$$ \frac{a^n}{a^n } = a^{n-n} = a^0 \quad y \quad \frac{a^n}{a^n } = 1 \quad luego \quad a^0 = 1$$
</font>

<font color = "blue">
## Potencias con exponente entero negativo 

$$a^n \cdot a^{-n} = a^{n+(-a)} = 1 \cdot$$ $$\cdot a^{-n} = \frac {1}{a^n}$$

**Una potencia de exponente entero negativo es la inversa de la potencia con exponente entero positivo**.

$$Si \quad a ≠ 0  \Rightarrow  a^{-n} = \frac {1}{}a^n$$
    
</font>

<font color = "blue">
Para determinar el signo de una potencia con exponente entero debemos tener en cuenta  lo siguiente:

a).- Las potencias con exponente par son siempre positivas. Esto quiere decir que, si tenemos una potencia $a^b$, entonces:

* Si a es positivo y b es par, entonces $a^b$ es positivo.    $\quad 3³ = 3 \cdot 3 \cdot 3 = 27$

* Si a es negativo y b es par, entonces $a^b$ es positivo.    $\quad (-3)² = (-3) \cdot (-3) = 9
</font>

<font color = "blue">
b).- Las potencias con exponente impar siempre tienen el mismo signo que su base. Esto quiere decir que, si tenemos una potencia $a^b$, entonces:

* Si a es positivo y b es impar, entonces a^b es positivo.  $\quad 2³ = 2 \cdot 2 \cdot 2 = 8$

* Si a es negativo y b es impar, entonces a^b es negativo.   $\quad (-3^3) = (-3) \cdot (-3) \cdot (-3) = -27$
</font>

## Raíz cuadrada
<font color = "blue">
    
La raíz cuadrada de un número real a es otro número real b tal que al elevarlo al cuadrado se obtiene a.

Si $ a \geq  0, \quad \sqrt{a}= b  \Rightarrow   b^2=a$

El símbolo $\sqrt{a}$, se lee radical de índice 2 de a.
    
</font>

## Radicales de cualquier índice

<font color = "blue">
    
La raíz enésima de un número a es un número b tal que al elevarlo a n se obtiene a.

$ \sqrt[n]{a} = b   \Rightarrow   b^n=a$

El símbolo n√a
se llama radical de índice n y radicando a.     

</font>

## Radicales equivalentes

<font color = "blue">
    
Dos radicales son equivalentes cuando tienen la misma raíz.

### Suma y resta de radicales semejantes

<font color ="blue">
* Dos radicales son semejantes si tienen el mismo índice y el mismo radicando.

* Para sumar o restar radicales semejantes, se extrae la raíz como factor común y se operan los coeficientes.
    
$$a \sqrt[n]{c} + b \sqrt[n]{c} = (a+b)\sqrt[n]{c}$$
    
    $$a \sqrt[n]{c} - b \sqrt[n]{c} = (a-b)\sqrt[n]{c}$$
    
</font>

## Potencias de exponente 

Si a > 0, se define potencia de base a y exponente fraccionario como: 

$$ a^{\frac{m}{n}} = \sqrt[n]{m}$$


* **Denominador**: índice del radical.
* **Numerador**: exponente del radicando.
