## Teorema de pitágoras

$h^2 = a^2 + b^2 $  $\quad h = \sqrt{a^2 + b^2}$

$ a^2 = h^2 - b^2$ $\quad a = \sqrt{h^2-b^2}$

$b^2 = h^2 - a^2$ $\quad b = \sqrt{h^2 - a^2}$

# Calcular la hipotenusa "h" conociendo los lados "a" y "b"
import math

a = 3

b = 4

h = math.sqrt(a**2 + b**2)

print ("La hipotenusa mide", h)

# Calcular uno de los lados, conociendo el otro y la hipotenusa
import math
h = 5

a = 3

b = math.sqrt(h**2 - a**2)

print ("El lado {} mide {}".format("'a'", b), "cm^2")
