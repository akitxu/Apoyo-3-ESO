<font color ="blue"> **Tercero**. Se resuelven las sumas y restas en el mismo orden. </fonrt>

<font color ="red"> $8x^2-3x^2+5x^2 = 10x^2$ </fonrt>

**<font color ="blue"> ¿Cómo se expresa un polinomio de la forma $a^2-b^2$ como producto de una suma por una diferencia?.</fonrt>**

**<font color ="blue">Ejemplo: Expresa $P(x) = 16-x^2 )$ como el producto de una suma por una diferencia </fonrt>

<font color ="blue">**Primero. Se identifican **a** y **b****  </fonrt>

<font color ="blue"> $a^2 = 16 \rightarrow  a =\sqrt{ 16 } = 4$ </fonrt>

<font color ="blue"> $b^2 = x^2 \rightarrow  a =\sqrt{ x^2 } = x$ </fonrt>

<font color ="blue">**Segundo. Se alica la igualdad**   </fonrt>

<font color ="blue">$a^2-b^2 = (a+b)(a-b) </fonrt>

<font color ="blue"> $16-x^2 = 4^2-x^2 = (4+x)(4-x) </fonrt>
