# <font color ="blue">Expresión algebraica</fonrt>
 
 <font color ="blue">**Expresión algebraica** es una expresión matemática en la que intervienen letras, números y los signos de las operaciones aritméticas.</fonrt>
 
 <font color ="blue"> Las letras reciben el nombre de **variables** o **incógnitas** y representan números o cantidades desconocidas.</fonrt>


# <font color ="blue">  Valor numérico </fonrt>

<font color ="blue"> **Valor numérico** de una expresión algebraica es el resultado obtenido al sustituir cada una de las variables por un número determinado.  </fonrt>

# <font color ="blue"> Monomio  </fonrt>

<font color ="blue"> **Monomio** es una expresión algebraica formada por el producto de un número por una o varias variables elevadas a exponentes naturales.  </fonrt>

<font color ="blue"> Los monomios constan de dos partes:  </fonrt>

<font color ="blue">  - **Parte literal**: formada por las variables y sus exponentes correspondientes. </fonrt>

<font color ="blue"> - **Coeficiente**: o parte numérica, es el número que multiplica a la parte literal.  </fonrt>

<font color ="blue"> **Grado de un monomio** es la suma de todos los exponentes de las variables que forman la parte literal.  </fonrt>

# Polinomios

<font color ="blue"> **Polinomio** es una expresión algebraica formada por la suma de varios monomios no semejantes, que se llaman términos.  </fonrt>

<font color ="blue"> - El término principal es el monomio de mayor grado. Su coeficiente es el coeficiente principal.  </fonrt>

<font color ="blue"> - El término independiente es el monomio que no tiene parte literal.  </fonrt>

<font color ="blue"> - El grado de un polinomio es el mayor de los grados de los monomios que lo forman.  </fonrt>

# Identidades notables

<font color ="blue"> **Cuadrado de una suma** El cuadrado de una suma de dos términos es igual al cuadrado del primer término más el cuadrado del segundo término más el doble del primero por el segundo.
$$(a + b)^2 = a^2 + b^2 + 2ab)$$  </fonrt>



<font color ="blue">  **Cuadrado de una diferencia** El cuadrado de una diferencia de dos términos es igual al cuadrado del primer término más el cuadrado del segundo término menos el doble del primero por el segundo.
$$ (a-b)^2 = a^2 + b^2 -2ab$$ </fonrt>

<font color ="blue">**Suma por diferencia** La suma de dos términos por su diferencia es igual al cuadrado del primer término menos el cuadrado del segundo término.
$(a+b)(a-b) = a^2 - b^2$ </fonrt>
