    • Entender el concepto de elemento y mezcla a partir de la teoría de Dalton.

## Mezclas y sustancias puras

Una mezcla es una combinación de varias sustancias en proporción variable que se pueden separar mediante procedimientos físicos.

### Mezclas heterogéneas y homogéneas

#### Mezclas heterogéneas

La **mezcla heterogénea** **está  formado  por  dos  o  más sustancias**, y las propiedades difieren de un punto a otro del sistema. Además, **a simple vista se pueden distinguir los  componentes**.  Por  ejemplo:  una  mezcla  de  arena  y  agua  o  una  roca  de  granito.

#### Mezclas homogéneas

La mezcla **homogénea es uniforme**, y **las propiedades son las mismas en cualquier parte de la mezcla**. En las mezclas homogéneas **no se distinguen sus componentes a simple vista**.

#### Sustancia pura

Una sustancia pura es un tipo de materia que tiene una composición constante y unas propiedades características.

**Sustancias simples:** No pueden descomponerse en sustancias más sencillas. Por ejemplo, el hidrógeno y el oxígeno.

**Sustancias compuestas:** Se pueden descomponerse en otras más sencillas mediante procedimientos químicos.


Razona cuáles de los siguientes materiales están formados por sustancias puras y cuáles por mezclas.

    * Agua mineral con gas
    * Anillo de acero
    * Papel de celulosa pura
    * Mosto de uva
    * Oxígeno medicinal 100 %



#### De sustancia compuesta a sustancia simple. Procedimientos

* **Descomposición térmica**. Mediante el calentamiento de una sustancia compuesta se produce la descomposición. 

* **Electrolisis**. se produce por el paso de la corriente eléctrica a través de una sustancia, fundida o disuelta.

#### Diferencia entre compuesto y mezcla.

* **Mezcla:** Los constituyentes de una mezcla pueden estar en cualquier proporción y mantienen sus propiedades características. Ejemplo grabanzos y agua. 

* **Compuesto:** Los elementos constituyentes de un compuesto están siempre en la misma proporción y la sustancia final es una nueva sustancia pura con otras propiedades. Ejemplo el agua.

Los compuestos, a diferencia de las mezclas, solo pueden separarse en sus componentes mediante procedimientos químicos.

## Mezclas heterogéneas

### Filtración

La **filtración** se utiliza para separar una mezcla heterogénea formada por un sólido insoluble en un líquido.

### Decantación

La **decantación** es un método general para separar una mezcla heterogénea, formada por un sólido o un líquido más denso, de un líquido menos denso.

### Separación mgnética

La **separación magnética** es una técnica para separar sustancias que son atraídas por un imán (sustancias ferromagnéticas), como el hierro, el acero y el níquel, de otras que no lo son.

## Centrifugación

La **centrifugación** es una técnica empleada para acelerar la decantación y separar rápidamente mezclas heterogéneas (sólido-líquido) cuando sus componentes tienen diferente densidad.

## Mezclas homogéneas

Una **disolución es una mezcla homogénea de dos o más componentes en proporción variable**. 

**Disolvente** Es el componente que se encuentra en mayor proporción 

**Soluto** Es el componente que se encuentra en menor proporción.

### Cristalización

La **cristalización** se emplea para obtener cristales de un soluto sólido disuelto en una disolución concentrada, evaporando el disolvente hasta que el soluto cristalice. Ejemplo las salinas de agua de mar.

### Destilación

La **destilación** es la separación por acción del calor de un líquido y de un sólido (o de otro líquido con diferente punto de ebullición).   

## Mezclas de especial interés

### Aleaciones

Una **aleación** es una mezcla homogénea formada, generalmente, por dos o más metales. Las aleaciones tienen como objetivo mejorar las propiedades de los metales que las integran.

**Acero: Aleación de hierro y carbono.

**Bronce:** Aleación formada por cobre y estaño.

**Latón:**  Aleación de cobre y cinc con un color amarillo brillante.


**Acero:**  Aleación de hierro con un pequeño porcentaje de carbono. El acero inoxidable, además, contiene cromo y níquel para hacerlo más resistente a la oxidación y mantenerlo brillante y resistente.

### Coloides

Un **coloide** es una dispersión de partículas de una sustancia, la **fase dispersa,** entre las partículas de otra, el **medio de dispersión**.

## Concentración de una disolución

La concentración de una disolución expresa la proporción entre el soluto y el disolvente e indica la **cantidad de soluto que hay disuelto en una determinada cantidad de disolución**.


### Porcentaje en masa

El **porcentaje en masa** se define como la **masa de soluto (en gramos) disuelta en 100 g de disolución**. Se calcula mediante la relación:

$$Porcentaje \; en \; masa = \frac{Masa \; de \; soluto  \; (g)}{Masa \; de \; disolución \; (g)} \cdot 100$$

## Porcentaje en volúmen

El **porcentaje en volumen** se define como el **volumen de soluto (en mL) disuelto en 100 mL de disolución**. Se calcula mediante la relación:

$$Porcentaje \; en \; volumen = \frac{Volumen \; de \; soluto \; (mL) }{ Volumen \; de \; disolución \; (mL)}        \cdot 100$$

## Concentración en masa

La **concentración en masa** se define como la **masa de soluto disuelta en un litro de disolución**. Se calcula mediante la relación:

$$Concentración \; en \; masa =       \frac{Masa \; de \; soluto \; (g)  }{ Volumen \; de \; disolución \; (L)} $$

### Solubilidad e identificación de sustancias.

La solubilidad es una propiedad característica de las sustancias. Por ello se utiliza en la identificación de sustancias desconocidas gracias a una tabla de solubilidades.

### Solubilidad y temperatura.

En general, **la solubilidad de los sólidos y líquidos aumenta con la temperatura, mientras que la de los gases disminuye**.
