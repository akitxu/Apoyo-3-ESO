**Instrucciones**

1.- Abrir esta página Web: [link](https://hub.gke.mybinder.org/user/akitxu-apoyo-3-eso-upnr35cq/tree)
 
2.- pulsar con el ratón en **Binder home page**

3.- En la ventana Github introducir  https://github.com/akitxu/Apoyo-3-ESO

4.- Pular en **launch** y esperar a que se abra MyBinder.

En la ventana que se abre podemos apreciar varias cosas.

* En la parte superior derecha hay cuatro pestañas, Con la pestaña "Join this repo´s Video Chat" podemos inicair un video chat con los alumnos.

* Vemos diversas carpetas a las que podemos acceder, "Fisica y Química 3º ESO", "Matemáticas 3º ESO", etc.

* Entrando en alguna de estas carpetas, accedemos a los notebooks desarrollados de éste tema. Son ficheros cuya extensión es ".ipynb"

* Abramos, por ejemplo el notebook "2ºESO-- Expresiones Algebraicas - 2_rev1 .ipynb"

* En la parte sueprior de la ventana que se abre con la teoría y ejercicios, hay varias pestañas. 

        * La pestaña Download descarga este fichero .ipynb  con este formato.
        * En la pestaña "File", y "Download as", podemos descargar este fichero en diferentes formatos, HTML, pdf, etc.

Estoo nos sirve para que los alumnos, una vez realizados sus ejericios en este notebook, lo descarguen y lo envíen, vía @mail, por ejmplo, a su profesor, aguelo tutor, etc.

Si nos fijamos en los ejercicios, vemos que están escritos en celdas de este libro. Las celdas pueden tener varios formatos. Nosotros utilizaremos el formato Markdown para escribir testo, y el formato "code", para resolver con Python , algunos problemas que requieran, por ejmplo, la aplicación del teorema de pitágoras, utilizando esa celda en lugar de una claculadora, con lo que lo vamos oerientando poco a cpoco hacia la programación. 

Para poner la celda en el modo escritua, pulsamos unavez sobre ella con el botón izdo. del ratón.

Para pasar a su estado de lectura, pulsamos la pestaña "Run"

Con la celda en el modo escritura, el alumno puede desarrollar todo el ejercicio como si lo estuviera haciendo con un bolígrafo sobre papél, ó simplemente poner su solución final, para después descargar el notebook y enviarlo para su corrección.

Poco a poco se les puede ir enseñando, al mismo tiempo, cómo dar formato al texto utilizando [comandos markdown](https://markdown.es/sintaxis-markdown/). No obstante, aunque inicialmente escriban de forma normal, no pasa nada, ya que, seguramente se verá mucho mejor que lo que desarrollan en papel.

Con muy poco código Latex, fácil de aprender por estos alumnos de 14 años, pueden desarrollar sus fórmulas en este libro.
Sentencias mínimas para iniciarse son:

* Las fórmulas empiezan y acaban con el símbolo $

* Para que se muestre centrada en la celda, la empezamos y finalizamos con dos symbolos $$

* Subíndices:   a1    Latex  $a_1$

* Subíndice con dos dígitos o letrs:  a12  latex  $a_{12}$

* Elevar al cuadrado:  a^2 ó a**2  latex $a^2$

* Raíz cuadrada de un número:        latex $\sqrt {345}$

* Raíz enésima de un número:      latex  $\sqrt{n}{567}$

* Punto utilizado como signo de multiplicación: 3.2 = 6      latex $3 \cdot 2 = 6$

* X utilizada como signo de multiplicación: 3 x 2 = 6       latex $ 3 \times 2 = 6$
