## Listar números primos

**nmax** variable para determinar la cantidad total de números primos que queremos listar.

```python
primos = [2]
nmax = 30
for x in range(3, nmax):
    for i in range(2, x):
        if x%i != 0:
            # Si i no es divisor de x, x puede ser primo
            continue
        else:
            # Si i es divisor de x, x no es primo
            break #No es necesario buscar más divisores
    else:
        # El bucle for ha terminado con normalidad
        # El número que estábamos comprobando es primo
        print (f"El número {x} es primo")

        primos.append(x)
F = open('numerosprimos.txt', 'w')
for data in primos:
    F.write('%d\n'%data)
```

## Teorema de pitágoras

```python
from math import sqrt
try:
    input = raw_input
except NameError:
    pass
def get_numeric_value(prompt):
    """
    Solicita un valor numérico de la entrada estándar.
    Devuelve None si no se introduce ningún valor.
    """
    while True:
        value = input(prompt)
        if not value:
            return None
        try:
            value = int(value)
        except ValueError:
            print("Debe ingresar un valor numerico. "
                      "Vuelva a intentarlo.")
        else:
            break
    return value

def get_triangle(a=None, b=None, h=None):
    """
    Devuelve un triángulo formado con guiones, con los valores especificados..
    """
    return r"""
        |\
        | \
        |  \ h
      a |   \
        |    \
        |_ _ _\
          b
        
    a = %s
    b = %s
    h = %s""" % ("?" if a is None else a,
                     "?" if b is None else b,
                     "?" if h is None else h)
def main():
    print(get_triangle())
    print("\nDeje en blanco el valor que desconoce.\n")
        
    a = get_numeric_value("a: ")
    b = get_numeric_value("b: ")
        
    if a is None or b is None:
        h = get_numeric_value("h: ")
    else:
        h = None
        
    if (bool(a) + bool(b) + bool(h)) < 2:
        print("Debe especificar, al menos, dos valores.")
        return 0
        
    if h is None:
        h = sqrt(a**2 + b**2)
    elif a is None:
        a = sqrt(h**2 - b**2)
    elif b is None:
        b = sqrt(h**2 - a**2)
        
    print("")
    print(get_triangle(a, b, h))
        
    return 0
if __name__ == '__main__':
    main()
```

```python
import math
def trapecio(x):
    B = float(input("Introduzca el valor de la base menor: "))
    b = float(input("\nIntroduzca el valor de la base mayor: "))
    h = float(input("\nIntroduzca el valor de la altura: "))
    A = ((B + b) / 2 ) * h

    print(f"El area del trapecio es {A}")

trapecio(0)
```

## Area y volúmen del cubo
 
 $A = 6 \cdot a² \quad \quad V = a³$

```python
import math
# Area y volúmen del cubo
def cubo(x):
    a = float(input("Introduzca el valor del lado: "))
    A = 6*(a**2)
    V = (a**3)
    print("\nEl area del cubo es: ",A)
    print("\nEl volumen del cubo es: ",V,"\n")
    
cubo(0)
```

## Area y volúmen del cilindro
 
 $Al = 2 \cdot \pi \cdot r \cdot h \quad \quad At = 2\cdot \pi \cdot r (r+h) \quad \quad V = \pi \cdot r² \cdot h$

```python
import math
# Areas y volúmen del cilindro
def cilindro(x):
    r = float(input("\nIntroduzca el radio del cilindro "))
    h = float(input("\nIntroduzca la altura del cilindro "))
    Alat = 2*(math.pi)*r*h
    Atot = (2*(math.pi)*(h+r))
    V = (math.pi*(r**2)*h)
    print(f"\nEl area lateral del  cilindro es { Alat}")    
    print(f"\nEl area total del  cilindro es {Atot}")
    print(f"\nEl volumen del cilindro es {V}")
cilindro(0)
```

## Area y volúmen del ortoedro

$Al = 2 (a \cdot b + b \cdot c + a \cdot c) \quad \quad V = a \cdot b \cdot c$

```python
# Area y volúmen del ortoedro
import math
def ortoedro(x):
    a = float(input("Introduzca el valor del ancho: "))
    b = float(input("\nIntroduzca el valor del largo: "))
    c = float(input("\nintroduzca el valor de la altura: "))
    Al = 2*((a*b)+(b*c)+(a*c))
    V = a*b*c
    print(f"\nEl area lareral del Ortoedro es {Al}")
    print(f"\nEl volumen del ortoedro es {V}")
ortoedro(0)
```

## Area y volúmen del cono

$Al = \pi \cdot r \cdot g \quad \quad g = \sqrt{h² + r²} \quad \quad At = \pi \cdot r (r + g) \quad \quad V = \frac {1}{3} \cdot \pi \cdot r² \cdot h$

```python
# Area y volúmen del cono
import math
def cono(x):
    g = float(input("Introduzca la generatriz del cono"))
    h = float(input("Introduzca la altura del cono"))
    r = float(input("introduzca el radio del cono"))
    Al = math.pi * r * g
    At = (math.pi*r)*(g+r)
    V = (math.pi*(r**2)*h)/3
    
    print(f"\nEl area lateral del cono es {Al}")    
    print(f"\nEl area total del cono es {At}")
    print(f"\nEl volumen del cono es {V}")
cono(0)
```

## Area y volúmen del tronco de cono

$ A_l = \pi \cdot (R + r) \cdot g \quad \quad A_t = \pi \cdot [(R + r) \cdot g + R² + r²] \quad \quad V = \frac {1}{3}  \cdot \pi \cdot h (R² + r² + R \cdot r)$

```python
# Area y volúmen del tronco de cono
import math
def tronco(x):
    h = float(input("Introduzca la altura\n"))
    r = float(input("Introduzca el Radio Menor"))
    R = float(input("Introduzca el Radio Mauyor"))
    g = float(input("Introduzca la generatriz"))
    Al = math.pi *(R*r)*g
    At = math.pi*[(R+r) * g + R**2 +r**2]
    V = (math.pi*h*((R**2)+(r**2)+(R*r)))/3
    print(f"\nEl volumen del tronco de cono es: {V}")
    print(f"\nEl area lateral del tronco de cono es: {Al}")
    print(f"\nEl area total del tronco de cono es: {At}")    
    
tronco(0)
```

## Area y volúmen de la pirámide

$A_l = \frac {Perimetro_{base} \cdot h_c}{2} \quad \quad A_t = A_l + A_{base} \quad \quad V = \frac {1}{3} \quad A_{base} \cdot h$

```python
# Area y volúmen de la pirámide
import math
def piramide(x):
    b = float(input("Introduzca el Apotema de la base: "))
    c = float(input("Introduzca el Apotema de base triangular: "))
    l = float(input("Introduzca el Lado de la base: "))
    h = float(input("Introduzca la altura: "))
    a = ((4*l)*(b+c))/2
    v = ((l**2)*h)/3
    print("El area es: ",a,"\n")
    print("El Volumen es: ",v,"\n")
piramide(0)
```

## Area y volúmen del casquete

$V = \pi \cdot h² (r - \frac {h}{3}) = \frac{\pi}{6} \cdot h(\frac{3 \cdot c²}{4} + h²)$
