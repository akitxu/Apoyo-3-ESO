# El movimiento

**Movimiento** El movimiento es el cambio de posición de un cuerpo a lo largo del tiempo.

**Posición** Posición es el lugar en el que se encuentra un cuerpo en un instante de tiempo determinado. 

# La trayectoria y el desplazamiento

**Trayectoria** La trayectoria es el camino que sigue un cuerpo en movimiento. 

**Desplazamiento** El desplazamiento de un cuerpo es la distancia en línea recta que hay entre el punto inicial y el final de su trayectoria.

# Velocidad y aceleración

## El tiempo

**Tiempo** El tiempo es una magnitud física con la que medimos la duración de un suceso o la separación entre varios.

## La velocidad

**Velocidad** La velocidad expresa el espacio recorrido por un cuerpo sobre una trayectoria en una unidad de tiempo.

Para calcular la velocidad media (**vm**) de un móvil debemos dividir el espacio recorrido (**e**) entre el tiempo empleado (**t**) en ese desplazamiento. 

```python
from IPython.display import Image
Image(filename="E:/Victor/Imagenes/Velocidad_media.jpg")
```

## La aceleración

**Velocidad instantánea** La velocidad instantánea es el valor que toma la velocidad en cada momento.

**Aceleración** La aceleración expresa la variación de la velocidad de un cuerpo en la unidad de tiempo.

**Aceleración media** La aceleración media (**am**) de un móvil se calcula dividiendo la variación de la velocidad entre el tiempo transcurrido.

**Movimiento rectilíneo uniforme** Un cuerpo sigue un movimiento rectilíneo uniforme si su trayectoria es una línea recta y se desplaza con velocidad constante.

**Movimiento rectilíneo uniformemente acelerado** El movimiento rectilíneo uniformemente acelerado lo tienen los objetos que se desplazan en línea recta y con aceleración constante.
