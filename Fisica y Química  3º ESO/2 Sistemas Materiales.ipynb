# Lección 2: Los Sistemas Materiales

## La materia y sus propiedades generales, masa y volúmen

**Materia:** es todo aquello que tiene **masa** y **volúmen**

**Masa:** La masa **m**, es la cantidad de materia que tiene un cuerpo. En el Sistema Internacional su unidad es el **kg**.  Se mide con la **balanza**

**Equivalencia de unidades de masa**

    $$1 \; 7(tonelada) = 1000 \:kg$$
    $$1 \; kg = 1000 \; g = 10^3 \;  gr$$
    $$1 \; g = 1000 \; \; mg = 10^3 mg$$

**VOLÚMEN:** El volúmen, es la cantidad de espacio que ocupa un cuerpo. Su unidad en el Sistema Internaional es el $\; m³$.  

Para medir el volúmen de los líquidos en el laboratior se emplean, probetas, pipetas o buretas, graduadas en **litros(L)** o **mililitros (ml)**

Otras unidades de medida para el volúmen:

$1 \; m³ = 1000 \; L$$
$$1 \; dm³ = 1 \; L$$
$$1 \; cm³ = 1 \; ml$$

Para medir el volúmen de los sólidos, se pueden emplear dos procedimientos: Método matemático (cálculo matemático de su volúmen) y Método de inmersión.

**Propiedades características de la materia:** $\textit{Son aquellas que permiten clasificar o identificar la materia}$, como por ejemplo, color, olor, sabor, densidad, punto de fusión, punto de ebullición, dureza, solubilidad, conductividad, ...

## La densidad, una propiedad característica de la materia.

La densidad relaciona la masa de una sustancia con el volúmen que ocupa. Su valor no depende de la cantida de materia. La densidad solamente dependen de la naturaleza de las sustancias que la componen. En el Sistema Internacional se mide en $kg/m³$.

La densidad nos permite diferenciar sustancias. No depende de la cantidad de materia. Por eso se dice que es una propiedad característica de la materia.

## Teoría cinético molecular (TCM).

Una teoría o modelo consiste en hacer una serie de suposiciones (hipótesis) de cómo están hechas las cosas o de cómo funcionana, a lo que llamamos interpretación teórica.

La **Teoría cinético molecular** es un modelo que usan los científicos para explicar los comportamientos de la materia. Se basa en las hipótesis siguientes:

    * Las sustancias están formadas por moléculas.

    * Entre molécula y molécula no hay nada, es decir existe vacío.

    * Las moléculas están en contínuo movimiento, aumentando la velociadad de estas cuando aumenta la temperatura.

Según esta teoría:

**Modelo del estado sólido según esta teoría** 

    * Las moléculas están juntas y ordenadas.

    * En contínuo movimiento de vibración alrededor de una posición fija.

    * Como consecuencia de esto los sólidos tienen forma y volúmen fijos.

    * Se dilatan y contraen al variar la temperatura.
    
    * No se pueden comprimir.

**Modelo del estado líquido según esta teoría** 

    * Las moléculas están más separadas pero desordenadas, en contínuo movimiento de vibración y rotación.

    * Esto es debido a que las fuerzas entre las moléculas  son más débiles que en el estado sólido.

    * Los líquidos tienen volumen propio (no varia), pero pueden fluir (deslizar unas sobre otras) y adoptar la forma del recipiente que los contiene.

    * No pueden contraerse y expandirse.
    
**Modelo del estado gaseoso según esta teoría** 

    * Las moléculas está muy separadas (las fuerzas de cohesión son casi nulas)  y en contínuo movimiento, por lo que chocan unas contra otras y con las paredes del recipiente que contiene el gas.

    *  Por esto decimos que los gases ejercen presión.

    *  Las partículas se mueven en línea recta, cambiando su dirección cuando chocan. 

    * El tiempo que dura un choque es mucho menor que el epacio de tiempo que transcurre enter dos choques sucesivos en una molécula.
    
    * La forma y el volumen de los gases son variables.
    
    * Se comprimen con facilidad y se expanden contínuamente.



## Cambios de estado

Los cambios de estado son procesos físicos que se producen cuando una sustancia pasa de su estado de agregación a otra diferente.

Si el proceso absorve calor  el proceso se denomina **fusión y evaporación**.

Si el proceso desprende calor el proceso se denomina **condensación y solidificación**.

La materia se presenta en la naturaleza en tres estados: **sólido**, **líquido** y **gaseoso**

* AL calentar suficientemente un sólido se convierte en líquido.

* Calentándolo más, se convierte en gas.

* También puede haber paso directo de sólido a gas. Se denomina **sublimación**. El yodo es un ejemplo.

## Punto de fusión y punto de ebullición

Son propiedades caracteístics de la materia.

**Punto de Fusión:**  Temperatura que permanece constante mientras el sistema cambia del estado sólido al estado gaseoso. Depende de la presión del sistema.

**Punto de ebullición:** Temperatura que permanece constante mientras el sistema cambia del estado líquido al estado líquido. Depende de la presión del sistema.

**Evaporación:** Cambio de líquido a gas que se produce en la superficie del líquido y a cualquier temperatura inferior a la de ebullición.

## Interpretación crítica de la temperatura, la presión y los cambios de estado, según la Teoría Cinético Molecular (TCM)

### La temperatura

La **temperatura absoluta** de unn gas es una medida de la energía cinética media de las partículas. 

**Cero absoluto** es la temperatura a la que las moléculas no tendrán energía cinética, es decir, no se mueven.



### La presión

La **presión de un gas** es el resultado de las colisiones de sus partículas entre ellas y con las paredes del recipiente que las contienen.



### Los cambios de estado

**Transformación de sólido a líquido:** 

    * Cuando se calienta una sustancia en estado sólido, aumenta la vibración de sus partículas, aumenta su energía cinética y , como consecuencia, aumenta la temperatura.
    
    * Al llegar a un punto llamado **punto de fusión** La temperatura permanece constante mientra quede alguna parte del sólido sin cambiar a líquido.
    
    * Esta energía es empleada por las partículas para vencer las fuerzas de cohesión que mantenía unidas las moléculas del sólido, para pasarlo de sólido a líquido.
    
**Transformación de líquido a gas:** 

    * Cuando se calienta una sustancia líquida, aumenta la velocidad del movimiento de sus partículas y su energía cinética. 
    
    * Al llegar a un punto llamado **punto de febullición** La temperatura permanece constante mientra quede alguna parte del líquido sin cambiar a vapor.
    
    * Esta energía es empleada para vencer las fuerzas de cohesión que mantenía unidas las moléculas del líquido.
    * Si seguimos calentando el gas, la temperatura volverá a aumentar, al igual que la velocidad y enrgía cinética de las partículas del gas.

## Leyes de los gases.

### Relación entre el volúmen y la temperatura. Ley de Boyle-Mariotte

**Ley de Boyle-Mariotte:** A temperatura constante, el volumen ocupado por una determinada masa de gas es inversamente proporcional a la presión. Es decir, el producto de la presión por el volúmen es constante.

$$ P \cdot V = constante$$

### Relación entre la presión y el volúmen. Ley de Charles

**Ley de Charles:** A presión constante, el volúmen es directamente proporcional a la temperatura absoluta del gas. Se expresa con la ecuación siguiente.

$$\frac{V}{T} = constante \Rightarrow  \frac{V_1}{T_1} = \frac{V_2}{T_2}$$


**Ejercicio:** La presión que soporta un gas es de 710 mm de Hg cuando se encuentra a 10 °C de temperatura
en un recipiente de 20 L. Se comprime el recipiente hasta que el volumen es de 15 L,
manteniéndose la presión constante. ¿Cuál es la temperatura final del gas?

Sol: 212,25 K

Estado inicial: P1 = 710 mmHg; V1 = 20 L; T1 = 10º C

Estado inicial: P2 = 710 mmHg; V2 = 15 L; T2 = x C

Como varian el V y la T, se cumple que $\frac{V}{T} = constante$

Convertimos las unidades teniendo en cuenta que  **1 atm = 760 mm Hg** y que **ºK = ºC + 273**

$P_1 = 710 \; mmHg \cdot \frac{1 \; atm}{760 \; mmHg} = 0,3 \: atm$

$T1 = 10 \; ºC + 273 = 283 \; ºK$

$T_2 = \frac{V_2 - T_1}{V_1} = \frac{15 \; L \cdot 283 \; K}{20 \; L} = 212,25 \; K$

### Relación entre la presión y la temperatura. Ley de Gay-Lusac

**Ley de Gay-Lusac:**

A volumen constante, la presión ejercidapor una determinada masa de gas es directamente proporcional a su temperatura.

$$\frac{P}{T} = constante \Rightarrow \frac{P_1}{T_1} = \frac{P_2}{T_2}$$
