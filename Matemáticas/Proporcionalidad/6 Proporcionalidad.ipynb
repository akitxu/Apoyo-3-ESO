# Leccion 6:  Proporcionalidad

## Conceptos

**Magnitud :** <font color="blue">**Es toda propiedad  que se pueda pesar o medir**</font>, por ejemplo, el peso, la masa, la longitud, el volúmen, el tiempo, etc. Todas estas son **magnitudes de sistemas físicos** 

**Razón :** <font color="blue">La razón entre dos cantidades (A y B) comparables es el cociente de éstas</font>. Se puede expresar como una fracción $\frac{A}{B}$, o como un número decimal o un entero.

Se lee **A es a B**. Al numerador de la fracción se le conoce como **antecedente** y al denominador como **consecuente**.

Ejemplo: Victor dedica cada día 8 horas a sus estudios y 2 horas a juegos e Internet.  La razón entre el número de horas que dedica al estudio y las que dedica al juego y a Internet es :: $\frac{8}{2} = 4$

Esto quiere decir que por cada hora que Victor dedica a jugar, dedica 4 a estudiar.

**Proporción:** <font color="blue">**Una proporción es una igualdad entre dos razones**</font>. Dadas dos razones $\frac{A}{B} \quad$ y  $\quad \frac{C}{D}$, tendríamos una proporción si  $\frac{A}{B} = \frac{C}{D}$

Esta propoción se lee: A es a B como C es a D.

A **A** y **D** se les denomina **extremos**

A **B** y **C** se les denomina **medios** 

<font color="blue">**En todas las proporciones el producto de los extremos es igual al producto de los medios.**</font>  por lo que tendríamos que $\quad A \cdot D = B \cdot C.$
