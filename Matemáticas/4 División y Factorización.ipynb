## División de monomios

<font color="blue">
    
Para dividir dos monomios se dividen sus coeficientes y sus partes literales.
    
Ejemplos
    
$\frac{12x³}{30x²}   = \frac{12}{30} \cdot  \frac{x³}{x²}=  \frac{2}{5} x$
    
$\frac{6x²y³ }{ 3x³y}= \frac{6 }{3 } \cdot \frac{ x²}{x³ } \cdot \frac{ y³}{y } = \frac{ 2y²}{x }$
     
</font>

## División de un polinomio  por un número

<font color ="blue">
    
Cuando dividimos un polinomio por un número, el resultado es otro polinomio que cumple las siguientes características :

* El polinomio resultante es del mismo grado que el polinomio que fue dividido.
* Sus coeficientes resultan de dividir cada uno de los coeficientes del polinomio entre el número
* Se dejan las mismas partes literales.
    
Ejemplo
    
$\frac{2x³-4x²+6x-2}{2} = \frac{2x³}{2} - \frac{4x²}{2} + \frac{6x}{2} - \frac{2}{2} = x³ - 2x² + x³ - 1$
    
</font>

### Raíces de un polinomio
<font color="blue">
Se dice que x = a es una raíz del polinomio P(x), si el valor numérico de P(x) para x = a es nulo, es decir, x = a es raíz de P(x) si P(a) = 0.    
    

Ejemplo

¿Los valores x = 2 y x = −1 son raíces de P(x) = x⁴ − 3x³ + 5x − 2?

Se calcula el valor numérico del polinomio en x = 2 y x = −1.

    P(2) = 24 − 3 ⋅ 23 + 5 ⋅ 2 − 2 = 16 − 3 ⋅ 8 + 10 − 2 = 16 − 24 + 10 − 2 = 0 ⇒
    ⇒ **x = 2 es raíz de P(x)**.
    P(−1) = (−1)4 − 3 ⋅ (−1)3 + 5 ⋅ (−1) − 2 = 1 − 3 ⋅ (−1) + 5 ⋅ (−1) − 2 = −3 ⇒
    ⇒ **x = −1 no es raíz de P(x)**.

El número de raíces de un polinomio depende de su grado.

**Un polinomio de grado n puede tener, como máximo, n raíces reales**.    
</font>

### Teorema del resto
<font color="blue">
El resto, R, de dividir un polinomio P(x) entre un binomio (x − a), coincide con el valor numérico del polinomio en x = a, es decir, R = P(a).    
    
Ejemplo

Al dividir el polinomio P(x) = 2x² + 2x − 3 por el binomio (x − 3), se obtiene el cociente C(x) = 2x + 8 y el resto R = 21.    
</font>

## Factorización
<font color="blue">
Un polinomio es irreducible si no se puede escribir como producto de dos o más polinomios de grado menor.

Factorizar un polinomio es expresarlo como producto de polinomios irreducibles.
             
</font>

### Factorización mediante extracción de factor común
<font color="blue">
Cuando todos los términos de un polinomio poseen un factor común, se puede comenzar su factorización extrayendo dicho factor común.
Ejemplos

P(x) = x⁴ − 3x² = x² (x² − 3)
    
Q(x) = 2x4⁴ − 4x³ = 2x³ (x − 2)
             
</font>



### Factorización utilizando las identidades notables
<font color="blue">
A veces, el polinomio que se va a factorizar es el desarrollo de una identidad notable.
Ejemplos

P(x) = x² − 10x + 25 = (x − 5)²
    
Q(x) = x2 − 16 = (x − 4)(x + 4)


             
</font>

### Fracciones algebraicas equivalentes
<font color="blue">
    
[link](https://www.superprof.es/apuntes/escolar/matematicas/algebra/polinomios/fracciones-algebraicas.html#tema_fracciones-algebraicas-equivalentes)
    
Dos fracciones algebraicas $\quad \frac{P(x)}{Q(x)} \; y \; \frac{R(x)}{S(x)} \quad$ son equivalentes, y lo representamos por: 
    
$$\frac{P(x)}{Q(x)} = \frac{R(x)}{S(x)}$$
    
si se verifica que  $\quad P(x) \cdot S(x) = Q(x) \cdot R(x)$

Comprobar si son equivalentes las fracciones algebraicas

$$\frac{x+2}{x²-4} \; y \; \frac{1}{x-2}$$
    
Verificamos que el producto de extremos sea igual al producto de medios

 

$(x+2)\cdot (x-2)   =   (1)\cdot (x^{2}-4)$
    
$ x²-4 = x²-4$

Ambos productos son iguales, por lo que concluimos que ambas fracciones son equivalentes
             
</font>

### Construcción de fracciones algebraicas equivalentes
<font color = "blue">
Dada una fracción algebraica, si multiplicamos el numerador y el denominador de dicha fracción por un mismo polinomio distinto de cero, la fracción algebraica resultante es equivalente a la dada.
    
Si multiplicamos el numerador y el denominador de la fracción algebraica $\frac{x+3}{x-3} \quad$ por $\quad x-1$, obtenemos una fracción equivalente.
    
$\frac{x+3}{x-3} = \frac{(x+3)(x-1)}{(x-3)(x-1}$
    
Para verificar que son equivalentes, efectuamos el producto de extremos por el producto de medios

(x+3)(x-3)(x-1) = (x+3)(x-1)(x-3)
</font>    

### Simplificación de fracciones algebraicas
<font color="blue">
Para simplificar una fracción algebraica se divide el numerador y el denominador de la fracción por un polinomio que sea factor común de ambos.

**Ejemplo1**: Simplificar la fracción     $\quad \frac{x²+4x+4}{x²-4}$
    
El numerador es un cuadrado perfecto: $\quad x²+4x+4 =  (x+2)²$
    
El denominador es una diferencia de cuadrados  $\quad (x²-4) = (x+2)(x-2)$
    
El factor común es (x+2)
    
$\quad \frac{x²+4x+4}{x²-4} = \frac{(x+2)²}{(x-2)(x+2) = \frac{x+2}{x-2}}$
    
**Ejemplo2**: Simplificar la fracción  $\quad \frac{7x²-7}{7x³-7} $  
    
Factorizamos el numerador, el cual posee un factor común y una diferencia de cuadrados
    
$7x²-7 = 7(x²-1) = 7(x+1)(x-1)$

###  Suma y resta
<font color="blue">
* Si** sus denominador son iguales**, se mantiene el mismo denominador y se suman o restan los numeradores.

**Ejemplo 1**: Dadas $\quadA(x)= \frac{x}{x+1}\quad$ y $\quad B(x)= \frac{2x−1}{ x+1} \quad$ calcula A(x) + B(x) y A(x) − B(x).

* $A(x)+B(x)= \frac{x}{x+1} + \frac{2x−1}{x+1} = \frac{x+2x−1}{x+1} = \frac {3x−1}{x+1}$
    
* $A(x)+B(x)= \frac{x}{x+1} - \frac{2x−1}{x+1} = \frac{x-(2x−1)}{x+1} = \frac {x-2x+1}{x+1} = \frac {-x+1}{x+1}$$    
    
* Si **tienen distinto denominador**, se buscan fracciones equivalentes a cada una que tengan el mismo denominador. Después se operan como en el caso de las fracciones algebraicas con el mismo denominador.

**Ejemplo 2** 
    
Dadas $\quad A(x)= \frac {x−1}{x+2} \quad$ y $\quad B(x)= \frac{x−1}{x+1} \quad$, calcula A(x) + B(x) y A(x) − B(x).
    
* $A(x) + B(x) = \frac{x-1}{x+2} + \frac{x-1}{x+1} =  \frac{(x-1)(x+1)}{(x+2)(x+1)} + \frac{(x-1)(x+2)}{(x+2)(x+1)} = \frac{x²-1+x²+x-2}{(x+2)(x+1)} = \frac{2x² + x - 3}{x²+3x+2}$

* $\quad A(x) - B(x) = \frac{x-1}{x+2} - \frac{x-1}{x+1} = \frac{(x-1)(x+1) - (x+2)(x-1)}{(x+2)(x+1)} = \frac{x²-1-(x²-x+2x-2)}{x²+x+2x+2} = \frac{x²-1-x²+x-2x+2}{x²+3x+2} =\frac{-x+1}{x²+3x+2}$

###  Multiplicación
<font color="blue">
Para multiplicar fracciones algebraicas se multiplican los numeradores y los denominadores.
    
$A(x)⋅B(x)= \frac{P(x)}{Q(x)}  \cdot \frac {R(x)}{S(x)} = \frac {P(x) \cdot R(x)} \frac {Q(x) \cdot S(x)}$
    

Ejemplo

Multiplica las fracciones algebraicas  $\quad A(x)= \frac {2x+3}{x+1} \quad$ y $\quad B(x) = \frac {x−3}{x+2}$

$A(x)⋅B(x) = \frac {2x+3}{x+1} \cdot \frac {x−3}{x+2} =   \frac {(2x+3)(x−3)}{(x+1)(x+2)} = \frac {2x²−6x+3x−9}{x²+2x+x+2} = \frac {2x²−3x−9}{x²+3x+2}$
             
</font>

###  División
<font color="blue">
Para dividir dos fracciones algebraicas se multiplica la primera fracción por la inversa de la segunda.

$ A(x) : B(x)= \frac {A(x)}{B(x)} = \frac {P(x)}{Q(x)} : \frac {R(x)}{S(x)} = \frac {P(x) \cdot S(x)}{ Q(x) \cdot R(x)}$

Ejemplo

Divide las fracciones algebraicas $\quad A(x)= \frac {2x−2}{2x−1} \quad$ y $\quad B(x)= \frac {2x+1}{x+3}$

$A(x):B(x) = \frac {2x−2}{2x−1} : \frac {2x+1}{x+3} = \frac {2x−2}{2x−1} \cdot  \frac {x+3}{2x+1} = \frac {(2x−2)(x+3)}{ (2x−1)(2x+1)} = \frac {2x²+6x−2x−6}{4x²−1}  =  \frac {2x²2+4x−6}{4x²−1}$
